### Longest Substring Without Repeating Characters

Given a string s, find the length of the longest substring without duplicate characters.

A substring is a contiguous sequence of characters within a string.

### Example 1:

Input: s = "zxyzxyz"

Output: 3
Explanation: The string "xyz" is the longest without duplicate characters.

### Example 2:

Input: s = "xxxx"

Output: 1

### Constraints:

- 0 <= s.length <= 1000
- s may consist of printable ASCII characters.



### Problem def
- classic leetcode, 
- given a string, find the longest substring wihtout repeating characters
- just return the len of the substring


- we only need to check till we get a repeated char, after that we can skip the rest
- for this scenario we use a technique called sliding window

- lets make sure our subarray always doesnt have duplicates, 
- so when we find a duplicate, we remove from left until we no longer have duplicates.
- how we wil know instantly if we have duplicate? using set.
- complexity : O(n)
- mem : O(n) - because of set.


```python
def lengthoflongestsubstring(s):
    charSet=set()
    # we are doing sliding window , so we need two pointers left pointer and right pointer
    l=0
    res=0
    # right pointer is going to go through every single character
    for r in range(len(s)):
        while s[r] in charSet:# when r is a duplicate, we have to update our window.we have to remove take the left char and remove it from our set
            charSet.remove(s[l])# this we do because we can also have same character multiple time continuously.
            l+=1
        charSet.add(s[r])
        res=max(res,r-l+1)
    return res


```

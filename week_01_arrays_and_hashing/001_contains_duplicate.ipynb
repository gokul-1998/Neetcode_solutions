# 001. Contains Duplicate


https://neetcode.io/problems/duplicate-integer/question?list=neetcode150


### Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

### Example 1:

Input: nums = [1, 2, 3, 3]

Output: true

### Example 2:

Input: nums = [1, 2, 3, 4]

Output: false

# Solution

```python
# solution

from typing import List



class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        return len(set(nums))!=len(nums)
```

# Approach 1:
    

- `len(set(nums)) is not equal to len(nums)` then there is a duplicate element in the array and we can return true else false

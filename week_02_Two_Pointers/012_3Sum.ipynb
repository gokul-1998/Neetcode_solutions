Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] where nums[i] + nums[j] + nums[k] == 0, and the indices i, j and k are all distinct.

The output should not contain any duplicate triplets. You may return the output and the triplets in any order.

Example 1:

Input: nums = [-1,0,1,2,-1,-4]

Output: [[-1,-1,2],[-1,0,1]]

Explanation:
```
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
```
The distinct triplets are [-1,0,1] and [-1,-1,2].

Example 2:

Input: nums = [0,1,1]

Output: []

Explanation: The only possible triplet does not sum up to 0.

Example 3:

Input: nums = [0,0,0]

Output: [[0,0,0]]

Explanation: The only possible triplet sums up to 0.

Constraints:

- 3 <= nums.length <= 1000
- -10^5 <= nums[i] <= 10^5

### Notes
    - the array can have duplicates
    - we need to find 3 numbers a,b,c => a+b+c  that sum to 0
    - we cannot have duplicates in our solution

### Sol 1:[dont use]
    - triple for loop
    - time complexity: O(n^3)
    - space complexity: O(n) for the output list
    - we need to eliminate the duplicates
    - lets sort the array first, then we can easily skip the duplicates
    - time complexity: O(n^3) + O(nlogn) for sorting => O(n^3)
### Sol 2:
    - if we sort the array and iterate, we can skip the duplicates,from the top iteration, and then we can fix one number and find the other two numbers that sum to -fixed number
    - then the problem reduces to two sum
    - we can solve two sum with hashmap
    - but if we do this via two sum II , having left right pointers, then we can solve it in O(n^2) time and O(1) space
    - if we have sum>0 , we need to decrease the sum, so we move the right pointer to left
    - if we have sum<0 , we need to increase the sum, so we move the left pointer to right
    - time complexity: O(n^2) + O(nlogn) for sorting => O(n^2)
    - there could be duplicates in pointers so we need to skip them


````python
from typing import List
class Solution:
    def threeSum(self,nums:List[int]):
        res=[]
        nums.sort()
        for i,a in enumerate(nums):
            if i>0 and a==nums[i-1]:# start from second iteration and if the prev number is same as the current number continue or skip it
                continue
            l,r=i+1,len(nums)-1 # so we fixed the first element as i,and so l will be i+1
            while l<r:# left and right can be equal ,so we dont put `l<=r`
                summ=a+nums[l]+nums[r]
                if summ>0:
                    r-=1
                elif summ<0:
                    l+=1
                else:
                    res.append((a,nums[l],nums[r]))
                    # after this we need to update our pointers, there are lots of way , be we will do the shortest with less code.
                    l+=1
                    while nums[l]==nums[l-1] and l<r:# note left pointer has to be less than right.if both pointers are equal , then its a collapse.[1,1,1,1,1] and we will get index out of bound error, this is to skip duplicates.
                        l+=1
        return res
````

### Common Pitfalls
1)  Forgetting to Skip Duplicates
A common mistake is not properly skipping duplicate values, which leads to duplicate triplets in the result. After sorting, when you find a valid triplet, you must skip over all identical values for the first element (outer loop) and the left pointer. Failing to do this causes wrong answers on inputs like [-1, -1, 0, 1, 1].

2) Not Sorting the Array First
The two-pointer approach only works correctly on a sorted array. Without sorting, moving pointers based on sum comparisons does not guarantee you will find all valid triplets. Always sort the input array before applying the two-pointer technique.

3) Incorrect Early Termination
When the first element nums[i] is positive, all remaining elements are also positive (since the array is sorted), so no triplet can sum to zero. However, incorrectly breaking when nums[i] >= 0 instead of nums[i] > 0 misses cases like [0, 0, 0]. The break condition should be nums[i] > 0, not nums[i] >= 0.

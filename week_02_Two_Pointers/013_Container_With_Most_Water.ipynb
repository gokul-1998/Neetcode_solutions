## Container With Most Water

- https://neetcode.io/problems/max-water-container/question?list=neetcode150

You are given an integer array heights where heights[i] represents the height of the 
ith bar.

You may choose any two bars to form a container. Return the maximum amount of water a container can store.

### Example 1:



### Input: height = [1,7,2,5,4,7,3,6]

Output: 36
### Example 2:

Input: height = [2,2,2]

Output: 4

### Constraints:

- 2 <= height.length <= 1000
- 0 <= height[i] <= 1000


### Question
    - we are given an array of heights, we should find what is the biggest area we can find with the container that has left height and right height.

### sol 1
    - we can check every l and r combination and get the max
    - the complexity will be O(n^2)


```python
# Brute force : two loops, compute all the area and get the max.
def maxArea(height):
    res=0 # op
    for l in range(len(height)): # left pointer
        for r in range(l+1,len(height)):
            area=(r-l)*min(height[l],height[r])
            res=max(res,area)
    return res
```

### sol 2 : O(n)

    - we know that the min heights are the limiters, so we will try a two pointer technique
    - l=0 , r=len(heights)-1
    - how to update the left and right pointers?
        - max_area=max_so_far
        - shift the pointer with less height, so that we increase the max_area_found_so_far
        - edge case, if both the pointers are same height, it doesnt matter which pointer you shift
        - terminating condition is l and right are at same point.

```python
def maxArea(height):
    res=0
    l,r=0,len(height)-1

    while l<r:
        area=(r-l)*min(height[l],height[r])
        if height[l]<height[r]:
            l+=1 # increase the one with lesser height.
        # elif height[l]>height[r]: this is not needed as both the lines do the same operation.
        #     r-=1
        else:# if both are equal, we can move any of the pointers
            r-=1
    return res
```

```python
class Solution:
    def maxArea(self, heights: List[int]) -> int:
        res=0
        l,r=0,len(heights)-1
        while l<r:
                area=(r-l)*min(heights[l],heights[r])
                res=max(area,res)
                # modify the one with lesser height
                if heights[l]<heights[r]:
                        l+=1
                elif heights[l]>heights[r]:
                        r-=1
                else:# when both are same we can move any pointer or both pointer
                        l+=1
                        r-=1
        return res
```

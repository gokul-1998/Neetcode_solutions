# 003. Two Sum


https://neetcode.io/problems/two-integer-sum/question?list=neetcode150


### Given an array of integers nums and an integer target, return the indices i and j such that nums[i] + nums[j] == target and i != j. You may assume that every input has exactly one pair of indices i and j that satisfy the condition.Return the answer with the smaller index first.

### Example 1:

Input: 
nums = [3,4,5,6], target = 7

Output: [0,1]
Explanation: nums[0] + nums[1] == 7, so we return [0, 1].

### Example 2:

Input: nums = [4,5,6], target = 10

Output: [0,2]
### Example 3:

Input: nums = [5,5], target = 10

Output: [0,1]

### Constraints:

- 2 <= nums.length <= 1000
- -10,000,000 <= nums[i] <= 10,000,000
- -10,000,000 <= target <= 10,000,000
- Only one valid answer exists.


# Solution

```python
from typing import List
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        seen={}
        for index,num in enumerate(nums):
            diff=target-num
            if diff in seen:
                return [seen[diff],index]
            seen[num]=index 
```

# Approach 1:
    

- create a dict seen to store 
- enumerate the nums and store the diff in a variable
    -  if the diff in seen , return `[seen[diff]`,i]
-  add the num as key and index as value
